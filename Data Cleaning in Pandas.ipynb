```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\bojja\Downloads\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns="Not_Useful_Column")
```

```python
df
```

```python
df.loc["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df.loc["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex = True)
```

```python
df.loc["Phone_Number"] = df["Phone_Number"].astype(str)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
import numpy as np

df["Phone_Number"] = df["Phone_Number"].replace(['nan--', 'Na--', 'N/a'], np.nan)
```

```python
df.loc["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df.loc["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df
```

```python
df.loc["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df.loc["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
df = df[df["Do_Not_Contact"] != 'Y']
df
```

```python
df = df.dropna(subset=["Phone_Number"])
df
```

```python
df = df.reset_index(drop=True)
```
